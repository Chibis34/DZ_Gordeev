# Задание 1. Базовое изучение

```python
import pandas as pd
import numpy as np
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv',
                header=None, names=["1. oper", "2. vozrast", "3. #bolnici", "temp_rekt", "pulce", 
                                    "chast_dihaniya", "7. temp kon", "8. perif imp", "9. sliz o", 
                                    "10. time zap kap", "11. bol'", "12. peristaltica", "13. vzdut jivot", 
                                    "14. naz gaz", "15. naz ref", "16. naz ref PH", "17. kal", "18. bruh pol", 
                                    "19. V yachey", "20. belok", "21. abdomin", "22. belok brusch", "23. rezult", 
                                    "24. hirurg povr", "25.", "26.", "27", "cp_data"], na_values = True
                 )
df2.head()
```

```python
df = df2.drop(["3. #bolnici", "7. temp kon",  "9. sliz o", "10. time zap kap", "11. bol'", "12. peristaltica", 
         "13. vzdut jivot", "14. naz gaz", "15. naz ref", "16. naz ref PH", "17. kal", "18. bruh pol", 
         "19. V yachey", "20. belok", "21. abdomin", "22. belok brusch", "25.", "26.", "27", "cp_data"]
                 , axis = 1)
```

```python
df[["temp_rekt", "pulce", "chast_dihaniya"]] = df[["temp_rekt", "pulce", "chast_dihaniya"]].apply(pd.to_numeric, errors='coerce')
df
```

## Базовые метрики

```python
print('Базовые метрики по тепературе: ')
print('Температура, max: ', max(df['temp_rekt']))
print('Температура, min: ', min(df['temp_rekt']))
print('Среднеарифметическое значение: ', df.temp_rekt.mean())
print('Медиана: ', df.temp_rekt.median())
print('Значение моды: ', df.temp_rekt.mode()[0])
print('СКО: ', df.temp_rekt.std())
print('Дисперсия: ', df.temp_rekt.var())
print('______________________________________________________________')      
print('Базовые метрики по пульсу:')
print('Пульс, max: ', np.max(df['pulce']))
print('Пульс, min: ', np.min(df['pulce']))
print('Среднеарифметическое значение: ', df.pulce.mean())
print('Медиана: ', df.pulce.median())
print('Значение моды: ', df.pulce.mode()[0])
print('СКО: ', df.pulce.std())
print('Дисперсия: ', df.pulce.var())
print('______________________________________________________________')       
print('Базовые метрики по частоте дыхания')
print('Частота дыхания, max', df['chast_dihaniya'].max())
print('Частота дыхания, min', df['chast_dihaniya'].min())
print('Среднеарифметическое значение: ', df.chast_dihaniya.mean())
print('Медиана: ', df.chast_dihaniya.median())
print('Значение моды: ', df.chast_dihaniya.mode()[0])
print('СКО: ', df.chast_dihaniya.std())
print('Дисперсия: ', df.chast_dihaniya.var())      
```

#### Согласно базовым метрикам имеется существенное различие между минимальным и максимальным значениями пульса и частоты дыхакния, и как следствие высокие показатели СКО и дисперсии. 
#### Минимальное и максимальное значение температуры не имеет столь существенного различия, что влияет на показатели СКО и дисперсии. Отсутствие существенного различие в данных обосновывается физиологией животных, т.к. слишком высокие/низкие показатели температуры приводят к смерти животного и соответственно нет необходимости/возможности измерения прочих параметров. В случае наличия большей выборки по температуре, то подобные минимальное/максимальное значения можно принять как пограничные (живое/мертвое животное).

# Задание 2. Работа с выбросами

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
df3 = df.drop(["1. oper", "2. vozrast", "8. perif imp", "23. rezult", "24. hirurg povr"]
                 , axis = 1)
print(df['temp_rekt'].mean()) # среднее
print(df[(df['temp_rekt'] > 35.7) & (df['temp_rekt'] < 39.5)]['temp_rekt'].mean())
print(df['temp_rekt'].median()) #медиана
print(df[(df['temp_rekt'] > 35.7) & (df['temp_rekt'] < 39.5)]['temp_rekt'].median())
print(df['temp_rekt'].round().mode()[0]) #мода
print(df[(df['temp_rekt'] > 35.7) & (df['temp_rekt'] < 39.5)]['temp_rekt'].round().mode()[0])
```

```python
q1 = df3['temp_rekt'].quantile(0.25)
q3 = df3['temp_rekt'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df3[df3['temp_rekt'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulce')
remove_outliers
```

```python
pd.concat([df3, remove_outliers]).drop_duplicates(keep=False)#.count()
```

#### Обнаружено 12 выбросов по температуре лошадей(менее 36.6 и более 39.6 ), при этом нормальная температура лошадей находится в диапазоне 37,5 -38,5 градусов.
#### В зависимости от цели данные выбросы можно исключить из выборки (при поиске иных более существенных факторов, влияющих на общее состояние лошади), не исключать (в целях понимания многофакторной взаимосвязи показателей), либо работать исключительно с данным выбросами паралельно анализируюя прочие параметры с отколонениями.

```python
print(remove_outliers['temp_rekt'].mean())
print(remove_outliers['temp_rekt'].median())
print(df3['temp_rekt'].mean() - remove_outliers['temp_rekt'].mean())
print(df3['temp_rekt'].median() - remove_outliers['temp_rekt'].median())
```

#### Исключение выбросов не целесообразно, т.к. возможно свокупность факторов(пульс, температура и пр. данные) может стать более показательной при определении резулбьатов исследования патологий животных.

```python
print(remove_outliers['pulce'].mean())
print(remove_outliers['pulce'].median())
print(df3['pulce'].mean() - remove_outliers1['pulce'].mean())
print(df3['pulce'].median() - remove_outliers1['pulce'].median())
```
