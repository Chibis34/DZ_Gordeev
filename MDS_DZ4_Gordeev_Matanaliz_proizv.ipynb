### Уровень 0:
Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

```python
import numpy as np
from sympy import *
from scipy.misc import derivative
```

```python
def f_1(x):
    return x ** 5 + 4 * sin(2 * x) + cos(3 * x + 3)
a = derivative(f_1, 1.0, dx=1e-6)
b = derivative(f_1, 1.0, dx=1e-6, n=2)
```

```python
print('Значение первой производной функции, в точке х = 1:', a)
print('Значение первой производной функции, в точке х = 1:', b)
```

### Уровень 1:
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
import matplotlib.pyplot as plt
```

```python
# Внесение значения функции
def f_2(x):
    return((sin((2.0 * x) + 1.0)) ** 5.0)
```

```python
item_x = np.linspace(-5, 5, num=50, endpoint=True, retstep=False, dtype=None)
```

```python
# График первой производной
plt.plot(item_x, [derivative(f_2, x, dx=1e-6) for x in item_x])
```

```python
# График второй производной
plt.plot(item_x, [derivative(f_2, x, dx=1e-6, n=2) for x in item_x])
```
