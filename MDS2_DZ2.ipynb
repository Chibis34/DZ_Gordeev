```python
import numpy as np
from scipy import linalg
```

Создайте случайную матрицу А1 размерностью 5 на 5

```python
A1 = np.array([[11,2,3,4,5],[6,7,80,9,10],[171,12,13,14,15],[116,17,18,19,209],[251,22,23,247,25]])  # создание матрицы
A1
```

● Транспонируйте матрицу A1

```python
np.transpose(A1)
```

● Найдите обратную матрицу для А1

```python
A2 = np.linalg.inv(A1) # обратная матрица
A2
```

```python
A2@A1 #проверка прошла
```

```python
A3 = np.dot(A2, A1) #проверка
A3
```

● Создайте диагональную матрицу - поэлементно перемножив единичную матрицу I и случайную матрицу A2 размерности 4 на 4

● Для всех полученных матриц найдите: определитель, ранг и собственные значения

```python
print("Матрица А1")
print('Определитель: ', np.linalg.det(A1))
print('Ранг: ', np.linalg.matrix_rank(A1))
print('Собственные значения: ', np.linalg.eigvals(A1))
print('______________________________________________________________________')
print("Матрица А2")
print('Определитель: ', np.linalg.det(A2))
print('Ранг: ', np.linalg.matrix_rank(A2))
print('Собственные значения: ', np.linalg.eigvals(A2))
print('______________________________________________________________________')
print("Матрица А3")
print('Определитель: ', np.linalg.det(A3))
print('Ранг: ', np.linalg.matrix_rank(A3))
print('Собственные значения: ', np.linalg.eigvals(A3))
print('______________________________________________________________________')
print("Матрица А4")
print('Определитель: ', np.linalg.det(A4))
print('Ранг: ', np.linalg.matrix_rank(A4))
print('Собственные значения: ', np.linalg.eigvals(A4))
print('______________________________________________________________________')
print("Матрица А5")
print('Определитель: ', np.linalg.det(A5))
print('Ранг: ', np.linalg.matrix_rank(A5))
print('Собственные значения: ', np.linalg.eigvals(A5))
print('______________________________________________________________________')
print("Матрица А6")
print('Определитель: ', np.linalg.det(A6))
print('Ранг: ', np.linalg.matrix_rank(A6))
print('Собственные значения: ', np.linalg.eigvals(A6))
```
